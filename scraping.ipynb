#### Extraction des avis sur TripAdvisor

```python
url = "https://www.tripadvisor.fr/Restaurant_Review-g187895-d25339020-Reviews-Osteria_Cancello_Dei_Macci-Florence_Tuscany.html"
driver = webdriver.Chrome()
driver.get(url)
wait = WebDriverWait(driver, 6)

cards = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div[data-test-target='review-body'] div.biGQs._P.VImYz.AWdfh")))

for card in cards:
    print(card.text)
    print("-----")

driver.quit()
```
