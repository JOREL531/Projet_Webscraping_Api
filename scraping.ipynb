-> Le ton de la r√©ponse d√©pend de la note, jsp si on fait comme √ßa. En l'occurence pour trustpilot, c'est gal√®re pour webscraper le rating donc √ßa r√©pond sur un ton 'formel'. 

```python
import json
from functions.response_generator import ResponseGenerator

# Init
generator = ResponseGenerator()

# Charger les avis
with open("reviews.json", 'r', encoding='utf-8') as f:
    liste_review = json.load(f)

print(f" {len(liste_review)} avis charg√©s\n")

# G√©n√©rer les r√©ponses
for i, review in enumerate(liste_review, 1):
    text = review.get('text', '') if isinstance(review, dict) else review
    rating = review.get('rating', None) if isinstance(review, dict) else None
    
    # D√©terminer le ton
    if rating:
        if rating >= 4:
            tone = "amical"
        elif rating <= 2:
            tone = "empathique"
        else:
            tone = "formel"
    else:
        tone = "formel"
    
    print(f"\nüìù AVIS #{i}")
    print("="*80)
    print(f" {text}")
    print(f"‚≠ê {rating}")
    print(f"\n R√âPONSE ({tone.upper()})")
    print("-"*80)
    print(generator.generate_response(text, tone=tone))
    print("="*80)
```
