## API

```python
app = FastAPI(
    title="Trustpilot Scraper API",
    description="API pour scraper des avis Trustpilot",
    # version="1.0.0"
)

@app.get("/reviews")
def get_reviews(url,  max_reviews:int):
    """
    Récupère les avis Trustpilot depuis une URL donnée
    """
    reviews = extract_review_from_trustpilot(url, max_reviews)

    return {
        "url": url,
        "requested_reviews": max_reviews,
        # "returned_reviews": len(reviews),
        "data": reviews
    }
```
