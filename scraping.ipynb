```python
def get_amazon_search(url):
    """Fonction pour obtenir le contenu de la page de l'élément que vous cherchez."""
    page = requests.get(url, headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def search_asin(asin):
    """Get the contents of a product pages using 'data-usin' number (unique identification number)."""
    url = "https://www.amazon.com/dp/"+asin
    page = requests.get(url, headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def search_reviews(review_link):
    """Passer le lien de "voir plus de commentaires" et extraire le contenu."""
    url = "https://www.amazon.com/"+review_link
    page=requests.get(url, headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```
